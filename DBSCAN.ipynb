```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=9,min_samples=3,metric='euclidean')
```

```python
x = d[['Annual_Income_(k$)', 'Spending_Score']]
```

```python
db.fit(x)
```

```python
db.labels_
```

```python
c=0
for i in db.labels_:
  if i==-1:
    c+=1
c
```

```python
import matplotlib.pyplot as plt
plt.scatter(x['Annual_Income_(k$)'],x['Spending_Score'],c=db.labels_)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.title('Customer Spending Score')
plt.show()
```
